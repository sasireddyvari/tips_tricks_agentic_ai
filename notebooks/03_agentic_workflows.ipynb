# Agentic Workflows

This notebook shows how to structure an agent as a sequence of stages instead of a single prompt.


## Why workflows matter

A graph-style workflow helps when you need:

- retries
- branching
- approvals
- human review


```python
state = {"input": "Investigate repeated timeouts after release"}

def gather_evidence(state):
    state["evidence"] = ["recent release", "timeout spike", "auth logs"]
    return state

def reason(state):
    state["decision"] = "escalate and rollback if release correlation is confirmed"
    return state

def review(state):
    state["approved"] = True
    return state

for step in (gather_evidence, reason, review):
    state = step(state)

state
```
