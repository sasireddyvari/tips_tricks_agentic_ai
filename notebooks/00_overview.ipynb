# Agentic AI Overview

This notebook explains the operating principles that show up in good agentic AI systems across domains.


## The core loop

A useful agent usually does four things well:

1. understand the request
2. gather evidence
3. choose a tool or next action
4. validate the result before returning it


```python
from dataclasses import dataclass

@dataclass
class AgentTip:
    name: str
    tip: str

tips = [
    AgentTip("small tools", "Keep each tool focused on one task."),
    AgentTip("evidence first", "Retrieve context before asking the model to reason."),
    AgentTip("structured output", "Return a typed result that downstream code can check."),
    AgentTip("fallbacks", "When confidence is weak, escalate instead of inventing details."),
]

tips
```

## Useful habits

- Separate retrieval from reasoning
- Keep prompts short and specific
- Log inputs, outputs, and latency
- Test the failure path as carefully as the happy path
