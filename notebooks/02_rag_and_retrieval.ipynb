# Retrieval-Augmented Generation

This notebook focuses on retrieval quality, citations, and grounding.


## Retrieval tips

- chunk by meaning, not only by size
- keep metadata with every chunk
- rank results before sending them to the model
- show citations in the final answer


```python
documents = [
    {"id": "doc-1", "text": "Authentication latency often comes from cache saturation around shift changes.", "topic": "identity"},
    {"id": "doc-2", "text": "Billing pipelines slow down when retries stack on top of month-end data skew.", "topic": "billing"},
    {"id": "doc-3", "text": "Alarm storms should be deduplicated and suppressed during approved maintenance windows.", "topic": "operations"},
]

query = "How do I handle login timeouts during shift change?"
matches = [doc for doc in documents if any(word in doc["text"].lower() for word in ["auth", "login", "timeout", "shift"])]
matches
```

```python
def build_answer(query: str, context: list[dict]) -> str:
    citations = ', '.join(item['id'] for item in context) if context else 'none'
    return f"Answer grounded in {citations}: check the auth cache, token validation latency, and the shift-change traffic pattern."

build_answer(query, matches)
```

## Tip

RAG is not just retrieval. The important part is making the answer traceable enough that a human can trust it.
