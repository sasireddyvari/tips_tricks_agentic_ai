# Evaluation and Guardrails

This notebook covers how to check whether an agent is actually safe and useful.


## What to measure

- task success
- tool accuracy
- citation quality
- latency
- cost per request
- human escalation rate


```python
metrics = {
    "success_rate": 0.92,
    "citation_accuracy": 0.89,
    "latency_ms": 840,
    "cost_per_request_usd": 0.004,
}
metrics
```

```python
def guardrail(answer: str, confidence: float) -> str:
    if confidence < 0.6:
        return "escalate to human review"
    if "maybe" in answer.lower():
        return "request more evidence"
    return "approve"

guardrail("The likely cause is cache saturation.", 0.81)
```

## Tip

A weak agent with good guardrails is better than a confident agent that cannot admit uncertainty.
