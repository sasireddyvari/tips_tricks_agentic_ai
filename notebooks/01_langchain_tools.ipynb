# LangChain Tools

This notebook shows practical patterns for creating tools that an agent can call.


## Tool design tips

Good tools are:

- deterministic when possible
- narrow in scope
- easy to validate
- safe to retry


```python
from typing import TypedDict

class ToolResult(TypedDict):
    name: str
    value: str
    confidence: float

def summarize_ticket(text: str) -> ToolResult:
    return {
        "name": "summarize_ticket",
        "value": text[:120].strip(),
        "confidence": 0.88,
    }

summarize_ticket("Customer is reporting login latency and repeated timeout errors during shift change.")
```

```python
# LangChain-style pattern: keep the tool interface stable and the implementation swappable.
# In a real app, this could wrap search, SQL, HTTP calls, or internal workflows.

def route_decision(signals: list[str]) -> str:
    if "billing" in signals:
        return "billing_playbook"
    if "auth" in signals:
        return "identity_playbook"
    return "general_triage"

route_decision(["auth", "timeout"])
```

## Tip

If you later swap the notebook demo into a live service, keep the tool signature stable and move the implementation behind it.
